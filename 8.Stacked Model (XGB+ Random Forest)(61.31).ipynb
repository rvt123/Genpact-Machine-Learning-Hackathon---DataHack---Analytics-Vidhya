```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("D://Analytics Vidya Genepact//train.csv")
meals=pd.read_csv("D://Analytics Vidya Genepact//meal_info.csv")
centre=pd.read_csv("D://Analytics Vidya Genepact//fulfilment_center_info.csv")
test=pd.read_csv("D://Analytics Vidya Genepact//test_QoiMO9B.csv")
```

```python
# On training Data 
train.set_index("meal_id",inplace=True)
meals.set_index("meal_id",inplace=True)
train=pd.concat([train,meals],join='inner',axis=1)

# On testing Data 
test.set_index("meal_id",inplace=True)
test=pd.concat([test,meals],join='inner',axis=1)
```

```python
train.head()
```

```python
test.head()
```

```python
# On training Data 
train.set_index('center_id',inplace=True)
centre.set_index("center_id",inplace=True)
train=pd.concat([train,centre],join='inner',axis=1)

# On Testing Data
test.set_index('center_id',inplace=True)
test=pd.concat([test,centre],join='inner',axis=1)
```

```python
train.reset_index(inplace=True,drop=True)
test.reset_index(inplace=True,drop=True)
```

```python
train.head()
```

```python
test.head()
```

```python
# An utility function to create dummy variable
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df
```

```python
train.set_index('id',inplace=True)
test.set_index('id',inplace=True)
```

```python
dummy=['emailer_for_promotion','homepage_featured','category','cuisine','city_code','region_code','center_type']
for var in dummy:
    train=create_dummies(train,var)
    test=create_dummies(test,var)
```

```python
test.head()
```

```python
train.drop('week',axis=1,inplace=True)
test.drop('week',axis=1,inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x

train.num_orders=outlier_capping(train.num_orders)
```

```python
import seaborn as sns
sns.distplot((train.num_orders))
```

```python
#Splitting the Columns
feature_columns = train.columns.difference( ['num_orders'] )
```

```python
from sklearn.cross_validation import train_test_split
from sklearn import metrics
```

```python
# Splitting the Data 
train_X, test_X, train_y, test_y = train_test_split( train[feature_columns],
                                                  train['num_orders'],
                                                  test_size = 0.3,
                                                  random_state = 123 )
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from xgboost import XGBRegressor
```

```python
cv_params = {'max_depth': [13,15], 'min_child_weight': [7,9]}
optimized_GBM = GridSearchCV(XGBRegressor(), 
                            cv_params, n_jobs = -1) 
```

```python
optimized_GBM.fit(train_X,train_y)
```

```python
optimized_GBM.best_params_
```

```python
xgb_13_7=XGBRegressor(max_depth=13,min_child_weight=7,n_jobs=-1,n_estimators=120)
xgb_13_7.fit(train_X,train_y)
xgb_13_9=XGBRegressor(max_depth=13,min_child_weight=9,n_jobs=-1,n_estimators=120)
xgb_13_9.fit(train_X,train_y)
```

```python
# Prediting with Random Forest 
xgb_pred_train = pd.DataFrame( { 'actual':  train_y,'predicted_13_7': xgb_13_7.predict( train_X ),'predicted_13_9':xgb_13_9.predict( train_X ) } )
```

```python
xgb_pred_train['predicted_xgb']=(xgb_pred_train['predicted_13_7']+xgb_pred_train['predicted_13_9'])/2
```

```python
xgb_pred_train.head()
```

```python
rmse_xgb_train = np.sqrt( metrics.mean_squared_error( train_y, xgb_pred_train.predicted_xgb) )
print (rmse_xgb_train)
print (metrics.r2_score( train_y, xgb_pred_train.predicted_xgb))
print()
```

```python
# Prediting with Random Forest 
xgb_pred_test = pd.DataFrame( { 'actual':  test_y,'predicted_13_7': xgb_13_7.predict( test_X ),'predicted_13_9':xgb_13_9.predict( test_X ) } )
```

```python
xgb_pred_test['predicted_xgb']=(xgb_pred_test['predicted_13_7']+xgb_pred_test['predicted_13_9'])/2
```

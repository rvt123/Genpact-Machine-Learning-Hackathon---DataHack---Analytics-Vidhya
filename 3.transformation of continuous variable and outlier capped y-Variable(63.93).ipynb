```python
for var in cont_var:
    train[var]=outlier_capping(train[var])
    test[var]=outlier_capping(test[var])
```

```python
squared = (train_cont**2).add_suffix("_squared")
square_root = (train_cont**0.5).add_suffix("_sqrt")
natural_log = np.log(train_cont+1).add_suffix("_ln")
train_cont= pd.concat([train_cont, squared, square_root,natural_log], axis=1)

squared = (test_cont**2).add_suffix("_squared")
square_root = (test_cont**0.5).add_suffix("_sqrt")
natural_log = np.log(test_cont+1).add_suffix("_ln")
test_cont= pd.concat([test_cont, squared, square_root,natural_log], axis=1)
```

```python
# Adding the the selected variable with their transformation in the list for further feature Generation 
train_cont_var= []
for variable in train_cont.columns:
     train_cont_var.append(variable)
        
# Adding the the selected variable with their transformation in the list for further feature Generation 
test_cont_var= []
for variable in test_cont.columns:
     test_cont_var.append(variable)
```

```python
print(train_cont_var)
print(test_cont_var)
```

```python
# Feature Generation by Taking every pair of numerical variable availabe and making variable by adding, subtracting,multiplying and dividing values 
for i in range(0,len(train_cont_var)):
    for j in range(i+1,len(train_cont_var)):
        train_cont['variableivariablej']= train_cont[train_cont_var[i]]+ train_cont[train_cont_var[j]]
        train_cont.rename(columns={'variableivariablej':train_cont_var[i]+"plus"+train_cont_var[j]},inplace=True)
        train_cont['variableivariablej']= train_cont[train_cont_var[i]]- train_cont[train_cont_var[j]]
        train_cont.rename(columns={'variableivariablej':train_cont_var[i]+"minus"+train_cont_var[j]},inplace=True)
        train_cont['variableivariablej']= train_cont[train_cont_var[i]]* train_cont[train_cont_var[j]]
        train_cont.rename(columns={'variableivariablej':train_cont_var[i]+"multiply"+train_cont_var[j]},inplace=True)
        train_cont['variableivariablej']= train_cont[train_cont_var[i]]/ train_cont[train_cont_var[j]]
        train_cont.rename(columns={'variableivariablej':train_cont_var[i]+"divide"+train_cont_var[j]},inplace=True)
```

```python
train_cont.head()
```

```python
# Feature Generation by Taking every pair of numerical variable availabe and making variable by adding, subtracting,multiplying and dividing values 
for i in range(0,len(test_cont_var)):
    for j in range(i+1,len(test_cont_var)):
        test_cont['variableivariablej']= test_cont[test_cont_var[i]]+ test_cont[test_cont_var[j]]
        test_cont.rename(columns={'variableivariablej':test_cont_var[i]+"plus"+test_cont_var[j]},inplace=True)
        test_cont['variableivariablej']= test_cont[test_cont_var[i]]- test_cont[test_cont_var[j]]
        test_cont.rename(columns={'variableivariablej':test_cont_var[i]+"minus"+test_cont_var[j]},inplace=True)
        test_cont['variableivariablej']= test_cont[test_cont_var[i]]* test_cont[test_cont_var[j]]
        test_cont.rename(columns={'variableivariablej':test_cont_var[i]+"multiply"+test_cont_var[j]},inplace=True)
        test_cont['variableivariablej']= test_cont[test_cont_var[i]]/ test_cont[test_cont_var[j]]
        test_cont.rename(columns={'variableivariablej':test_cont_var[i]+"divide"+test_cont_var[j]},inplace=True)
```

```python
# An utility function to create dummy variable
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df
```

```python
#Splitting the Columns
feature_columns = train.columns.difference( ['num_orders'] )
```

```python
from sklearn.cross_validation import train_test_split
```

```python
# Splitting the Data 
train_X, test_X, train_y, test_y = train_test_split( train[feature_columns],
                                                  train['num_orders'],
                                                  test_size = 0.3,
                                                  random_state = 123 )
```

```python
def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x
```

```python
train.num_orders=outlier_capping(train.num_orders)
```
